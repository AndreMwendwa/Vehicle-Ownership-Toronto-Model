Betas

```python
# Taken from Miller et. al (2005)
beta_aivtt = -0.05916234
beta_acost = 2.307699
beta_ivtt =  -0.05395848
beta_twalk = -0.07711395
beta_twait = -0.858
beta_tfare = 2.307699
```

```python
# Pivot aivtt to make 'Origin' the index and 'Destination' the columns
aivtt_pivot = aivtt.pivot(index='Origin', columns='Destination', values='Data')
aivtt_pivot
```

```python
acost_pivot = acost.pivot(index='Origin', columns='Destination', values='Data')
acost_pivot
```

```python
# Multiply by beta_aivtt, exponentiate, and take the log of the column sum
logsum_auto = np.log(np.exp(beta_aivtt * aivtt_pivot + beta_acost * acost_pivot).sum(axis=1))
logsum_auto
```

```python
tivtt_pivot = aivtt.pivot(index='Origin', columns='Destination', values='Data')
tivtt_pivot
```

```python
twalk_pivot = twalk.pivot(index='Origin', columns='Destination', values='Data')
twalk_pivot
```

```python
twait_pivot = twait.pivot(index='Origin', columns='Destination', values='Data')
twait_pivot
```

```python
tfare_pivot = tfare.pivot(index='Origin', columns='Destination', values='Data')
tfare_pivot
```

```python
logsum_transit = np.log(np.exp(beta_ivtt * tivtt_pivot + beta_twalk * twalk_pivot + beta_twait * twait_pivot + beta_tfare * tfare_pivot).sum(axis=1))
logsum_transit
```

```python
logsum_transit.fillna(-10e10, inplace=True)  # The trip is not possible
logsum_auto.fillna(-10e10, inplace=True)  # The trip is not possible
```

```python
def calculate_nearest_centroid(row, col_out):
    nearest_centroid = zones['Centroids'].distance(row.geometry).idxmin()
    return zones.loc[nearest_centroid, col_out]

data_missing['logsum_transit_alternative'] = data_missing.apply(calculate_nearest_centroid, col_out='logsum_transit', axis=1)
data_missing['logsum_auto_alternative'] = data_missing.apply(calculate_nearest_centroid, col_out='logsum_auto', axis=1)
```

```python
# Replace the nan values in logsum_transit and logsum_auto with the values in logsum_transit_alternative and logsum_auto_alternative
data = data.merge(data_missing[['Original_Index', 'logsum_transit_alternative', 'logsum_auto_alternative']], on='Original_Index', how='left')
data
```

Because there are some na values (individuals not in any GTHA zone), we'll use the logsum of the zone someone is in for those who are in a zone and the logsum of the nearest zone centroid for those who are not

```python
data['logsum_transit'] = np.where(data['logsum_transit'].isna(), data['logsum_transit_alternative'], data['logsum_transit'])
data['logsum_auto'] = np.where(data['logsum_auto'].isna(), data['logsum_auto_alternative'], data['logsum_auto'])
```

```python
sns.distplot(data['logsum_transit'])
```

The very tin and very large values of logsum transit are causing numerical problems. For this reason, we will bound it artificially between the values of 10 and 30, since that is where most values are found

```python
# Make all logsum_transit values < 10 equal to 10 and greater than 30 equal to 30
data['logsum_transit'] = np.where(data['logsum_transit'] < 10, 10, data['logsum_transit'])
data['logsum_transit'] = np.where(data['logsum_transit'] > 30, 30, data['logsum_transit'])
```
